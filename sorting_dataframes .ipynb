```python
import pandas as pd 
```

```python
#creating a dataframe 
df = pd.DataFrame({'col_1':['A','B','A','C','D'],
                   'col_2':[1,4,2,3,6],
                   'col_3':[12,3,1,22,1],
                   'col_4':['e','F','G','i','j']})
df
```

```python
#sorting dataframes in ascending order 
df.sort_values(by=['col_1'])
```

```python
#sorting in descending order 
df.sort_values(by=['col_1'], ascending=False)
```

```python
#sorting by multiple columns 

df.sort_values(by=['col_1','col_3'])
```

```python
#sorting by multiple columns 
df.sort_values(by=['col_1','col_2'])
```
